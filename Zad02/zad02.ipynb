## Zadanie 3: Narysować poniższy wykres

```python
fig, ax = plt.subplots()

angles = np.linspace(0, 2 * np.pi, 16, endpoint=False)  # 16 równomiernie rozmieszczonych kątów z zakresu 0 do 2π
x = 2 * np.cos(angles)  # cos(θ) daje współrzędną x punktu na okręgu
y = 2 * np.sin(angles)  # sin(θ) daje współrzędną y punktu na okręgu
ax.plot(np.append(x, x[0]), np.append(y, y[0]), 'r-', label='łamane', linewidth=2)

ax.plot([-1, 0, 1], [1, 0, 1], 'bd', label='punkty', markersize=10)

x_parabola = np.linspace(-1, 1, 100)
y_parabola = x_parabola ** 2 - 1
ax.plot(x_parabola, y_parabola, 'y-', label='sinus', linewidth=2)

ax.set_xlim([-2, 2])
ax.set_ylim([-3, 3])
ax.grid(True)
ax.legend()
plt.show()
```

## Zadanie 4: Punkty dla klasy Irys

```python
df = pd.read_csv('../Zad01/iris.txt', delimiter='\t')

label_map = {1: 'Setosa', 2: 'Versicolour', 3: 'Virginica'}
df['class'] = df['class'].map(label_map)

df
```

```python
virginica_samples = df[df['class'] == 'Virginica']
versicolour_samples = df[df['class'] == 'Versicolour']
setosa_samples = df[df['class'] == 'Setosa']
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0, 0].plot(virginica_samples['petal_length_in_cm'], virginica_samples['petal_width_in_cm'], 'r*', label='Virginica')
ax[0, 0].plot(versicolour_samples['petal_length_in_cm'], versicolour_samples['petal_width_in_cm'], 'gx', label='Versicolour')
ax[0, 0].plot(setosa_samples['petal_length_in_cm'], setosa_samples['petal_width_in_cm'], 'b+', label='Setosa')
ax[0, 0].set_xlabel('Atrybut 3 - petal length (cm)')
ax[0, 0].set_ylabel('Atrybut 4 - petal width (cm)')
ax[0, 0].legend()

ax[0, 1].plot(virginica_samples['sepal_width_in_cm'], virginica_samples['petal_width_in_cm'], 'r*', label='Virginica')
ax[0, 1].plot(versicolour_samples['sepal_width_in_cm'], versicolour_samples['petal_width_in_cm'], 'gx', label='Versicolour')
ax[0, 1].plot(setosa_samples['sepal_width_in_cm'], setosa_samples['petal_width_in_cm'], 'b+', label='Setosa')
ax[0, 1].set_xlabel('Atrybut 2 - sepal width (cm)')
ax[0, 1].set_ylabel('Atrybut 4 - petal width (cm)')
ax[0, 1].legend()

ax[1, 0].plot(virginica_samples['sepal_length_in_cm'], virginica_samples['petal_width_in_cm'], 'r*', label='Virginica')
ax[1, 0].plot(versicolour_samples['sepal_length_in_cm'], versicolour_samples['petal_width_in_cm'], 'gx', label='Versicolour')
ax[1, 0].plot(setosa_samples['sepal_length_in_cm'], setosa_samples['petal_width_in_cm'], 'b+', label='Setosa')
ax[1, 0].set_xlabel('Atrybut 1 - sepal length (cm)')
ax[1, 0].set_ylabel('Atrybut 4 - petal width (cm)')
ax[1, 0].legend()

ax[1, 1].plot(virginica_samples['sepal_width_in_cm'], virginica_samples['petal_length_in_cm'], 'r*', label='Virginica')
ax[1, 1].plot(versicolour_samples['sepal_width_in_cm'], versicolour_samples['petal_length_in_cm'], 'gx', label='Versicolour')
ax[1, 1].plot(setosa_samples['sepal_width_in_cm'], setosa_samples['petal_length_in_cm'], 'b+', label='Setosa')
ax[1, 1].set_xlabel('Atrybut 2 - sepal width (cm)')
ax[1, 1].set_ylabel('Atrybut 3 - petal length (cm)')
ax[1, 1].legend()

plt.tight_layout()
plt.show()
```
